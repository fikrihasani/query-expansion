```python
data_location = 'datacisi'
data_query = 'query.text' 
filename = data_location+'/'+data_query
filename
```

```python
data = []
with open(filename) as f:
    for line in f:
        data.append(line.strip("\n"))
            
f.close()
data
```

```python
while("" in data) : 
    data.remove("") 
data
```

```python
data[1437]
```

```python
querys = []
pos = 0
while pos < len(data)-1:
    tmp = ['','','','','']
    if ".I" in data[pos]:
        tmp[0] = data[pos]
        pos += 1
        if pos == len(data):
            break
    if ".T" in data[pos]:
        pos+=1
        string = ''
        while "." != data[pos][0]:
            string += " "+data[pos]
            pos+=1
        tmp[1] = string
        if pos == len(data):
            break
    if ".A" in data[pos]:
        pos+=1
        string = ''
        while "." != data[pos][0]:
            string += " "+data[pos]
            pos+=1
        tmp[2] = string
        if pos == len(data):
            break
    if ".W" in data[pos]:
        print("ada W")
        pos+=1
        string = ''
        while "." != data[pos][0]:
            string += ""+data[pos]
            pos+=1
        tmp[3] = string
        if pos == len(data):
            break
    if ".B" in data[pos]:
        pos+=1
        string = ''
        print(pos)
        while "." != data[pos][0]:
            string += " "+data[pos]
            pos+=1
            if pos == len(data):
                break
        tmp[4] = string
    querys.append(tmp)
print(querys)
```
